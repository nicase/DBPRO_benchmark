```python
import re
def convert_memory_to_mb(memory):
    matches = re.match(r'([\d.]+)(\w+)', memory)
    if matches:
        value = float(matches.group(1))
        unit = matches.group(2)
        if unit == "GiB":
            value = value * 1000
    else:
        print(memory)
    return value

logs_ann_glove_cosine["CPU Usage (%)"] = logs_ann_glove_cosine["CPU Usage (%)"].str.rstrip("%").astype(float)
logs_ann_glove_cosine["Memory usage (MB)"] = logs_ann_glove_cosine["Memory usage (MB)"].apply(convert_memory_to_mb)
```
